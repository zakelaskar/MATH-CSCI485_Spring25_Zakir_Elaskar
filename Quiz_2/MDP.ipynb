# Markov Decision Process (MDP)
----

**Value Iteration Process with Policy Changes in MDP**

We begin with a Markov Decision Process (MDP) where an agent decides whether to invest conservatively (C) or aggressively (A) in a financial portfolio. The objective is to find an optimal policy maximizing long-term rewards.

---

### **Step 1: Defining the MDP Components**

**States (S):**

- Low Wealth (L)
- Medium Wealth (M)
- High Wealth (H)

**Actions (A):**

- Conservative (C)
- Aggressive (A)

**Transition Probabilities:**

>| Current State | Action | Next State Probabilities     |
| ------------- | ------ | ---------------------------- |
| Low (L)       | C      | 80% Stay in L, 20% Move to M |
| Low (L)       | A      | 60% Stay in L, 40% Move to M |
| Medium (M)    | C      | 70% Stay in M, 30% Move to H |
| Medium (M)    | A      | 50% Stay in M, 50% Move to H |
| High (H)      | C      | 90% Stay in H, 10% Drop to M |
| High (H)      | A      | 70% Stay in H, 30% Drop to M |

**Rewards:**

- Low Wealth (L): -1
- Medium Wealth (M): 3
- High Wealth (H): 5

**Discount Factor (γ):** 0.9

---

### **Step 2: Value Iteration Updates**

We initialize values: $V_0(L) = 0$, $V_0(M) = 0$, $V_0(H) = 0$.

#### **Iteration 1**

Using Bellman’s equation:

>$
V_1(s) = \max_a \left[ R(s) + \gamma \sum_{s'} P(s' | s, a) V_0(s') \right]
$

For **Low Wealth (L):**

>$
V_1(L) = \max \left[ -1 + 0.9(0.8V_0(L) + 0.2V_0(M)), -1 + 0.9(0.6V_0(L) + 0.4V_0(M)) \right]
$

For **Medium Wealth (M):**

>$
V_1(M) = \max \left[ 3 + 0.9(0.7V_0(M) + 0.3V_0(H)), 3 + 0.9(0.5V_0(M) + 0.5V_0(H)) \right]
$

For **High Wealth (H):**

>$
V_1(H) = \max \left[ 5 + 0.9(0.9V_0(H) + 0.1V_0(M)), 5 + 0.9(0.7V_0(H) + 0.3V_0(M)) \right]
$

Since $V_0(L) = V_0(M) = V_0(H) = 0$, the initial values are just the rewards.

>$
V_1(L) = -1, \quad V_1(M) = 3, \quad V_1(H) = 5
$

#### **Policy Evaluation after Iteration 1**

> \$
Q(L, C) = -1 + 0.9(0.8(-1) + 0.2(3)) = -1.44
\$

> \$
Q(L, A) = -1 + 0.9(0.6(-1) + 0.4(3)) = -0.76
\$

> \$
Q(M, C) = 3 + 0.9(0.7(3) + 0.3(5)) = 4.92
\$

> \$
Q(M, A) = 3 + 0.9(0.5(3) + 0.5(5)) = 5.40
\$

> \$
Q(H, C) = 5 + 0.9(0.9(5) + 0.1(3)) = 8.82
\$

> \$
Q(H, A) = 5 + 0.9(0.7(5) + 0.3(3)) = 8.04
\$

**Policy at Iteration 1:**
- L → Conservative (C)
- M → Aggressive (A)
- H → Conservative (C)


#### **Iteration 2**

Updating $V_2(s)$:

>$
V_2(L) = \max \left[ -1 + 0.9(0.8(-1) + 0.2(3)), -1 + 0.9(0.6(-1) + 0.4(3)) \right]
$

>$
V_2(H) = \max \left[ 5 + 0.9(0.9(5) + 0.1(3)), 5 + 0.9(0.7(5) + 0.3(3)) \right]
$

Computing these:

>$
V_2(L) = -0.44, \quad V_2(M) = 5.46, \quad V_2(H) = 9.14
$

#### **Policy Evaluation after Iteration 2**

> \$
Q(L, C) = -1 + 0.9(0.8(-0.44) + 0.2(5.46)) = -0.1
>\$

> \$
Q(L, A) = -1 + 0.9(0.6(-0.44) + 0.4(5.46)) = 0.71
\$

> \$
Q(M, C) = 3 + 0.9(0.7(5.46) + 0.3(9.14)) = 6.93
\$

> \$
Q(M, A) = 3 + 0.9(0.5(5.46) + 0.5(9.14)) = 7.02
\$

> \$
Q(H, C) = 5 + 0.9(0.9(9.14) + 0.1(5.46)) = 11.29
\$

> \$
Q(H, A) = 5 + 0.9(0.7(9.14) + 0.3(5.46)) = 10.04
\$

**Policy at Iteration 2:**
- L → Aggressive (A)
- M → Aggressive (A)
- H → Conservative (C)

#### **Iteration 3**

Updating $V_3(s)$:

>$
V_3(L) = \max \left[ -1 + 0.9(0.8(-0.44) + 0.2(5.46)), -1 + 0.9(0.6(-0.44) + 0.4(5.46)) \right]
$

>$
V_3(M) = \max \left[ 3 + 0.9(0.7(5.46) + 0.3(9.14)), 3 + 0.9(0.5(5.46) + 0.5(9.14)) \right]
$

>$
V_3(H) = \max \left[ 5 + 0.9(0.9(9.14) + 0.1(5.46)), 5 + 0.9(0.7(9.14) + 0.3(5.46)) \right]
$

Computing these:

>$
V_3(L) = 0.81, \quad V_3(M) = 7.92, \quad V_3(H) = 11.58
$

#### **Policy Change Analysis**

From **Iteration 2 to Iteration 3**, let’s check the action values to determine if the policy changed.

For **Low Wealth (L):**

>$
Q(L, C) = -1 + 0.9(0.8(0.81) + 0.2(7.92)) = 0.43
$

>$
Q(L, A) = -1 + 0.9(0.6(0.81) + 0.4(7.92)) = 1.97
$

For **Medium Wealth (M):**

>$
Q(M, C) = 3 + 0.9(0.7(7.92) + 0.3(11.58)) = 8.72
$

>$
Q(M, A) = 3 + 0.9(0.5(7.92) + 0.5(11.58)) = 8.95
$

For **High Wealth (H):**

>$
Q(H, C) = 5 + 0.9(0.9(11.58) + 0.1(7.92)) = 13.06
$

>$
Q(H, A) = 5 + 0.9(0.7(11.58) + 0.3(7.92)) = 11.98
$

Since $Q(L, A) > Q(L, C)$ and $Q(H, C) > Q(H, A)$, the policy updates to:

- **Low Wealth (L)** → Aggressive (A)
- **Medium Wealth (M)** → Aggressive (A)
- **High Wealth (H)** → Conservative (C)



### Summary: Policy Evolution Over Iterations

>| State  | Iteration 1 | Iteration 2 | Iteration 3 |
|--------|------------|------------|------------|
| Low      | Conservative          | Aggressive            | Aggressive          |
| Medium   | Aggressive            | Aggressive            | Aggressive          |
| High     | Conservative          | Conservative          | Conservative          |

This table shows how the policy evolved over iterations. The action for Low Wealth (L) changed from Conservative (C) to Aggressive (A) in Iteration 2 and remained unchanged in Iteration 3. Medium Wealth (M) and High Wealth (H) retained their optimal actions after Iteration 1.
